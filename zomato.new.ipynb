```python
df["rate"].unique()
#from here we can see that it contains some values like nan, NEW,_ 
```

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value=str(value).split("/")#first i will split 
        value=value[0]#then i will take only that value which is at index 0
        return float(value)#now i am going to return float of this value

df2["rate"] = df2["rate"].apply(handlerate)
df2["rate"].head()
```

```python
#every value is containing commas that comma is making it string but cost should be string
def removecomma(value):
    value=str(value)
    if "," in value:
        value =value.replace(",", "")
        return float(value)
    else:
        return float(value)
    
df2["cost"]=df2["cost"].apply(removecomma)

    
```

```python
type=df2["type"].value_counts(ascending=True)
type
```

```python
type01=type[type<1000]
type01
```

```python
# now i want to replace the type(who are less than 1000) with others
def handle_rest_type(value):
    if( value in type01):
        return "others"
    else :
        return value
df2["type"]=df2["type"].apply(handle_rest_type)
```

```python
df2["location"]
```

```python
location=df2["location"].value_counts()
```

```python
#here we some locations which are very less in no., we have to replace them with other
location1=location[location<1000]
```

```python
def location0(value):
    if (value in location1):
        return "others"
    else:
        return value
df2["location"]=df2["location"].apply(location0)
```

```python
#now we will clean cuisines column
cuisines=df["cuisines"].value_counts()
```

```python
 cuisines1=cuisines[cuisines<100]
```

```python
def cuisines_less(value):
    if (value in cuisines1):
        return "others"
    else :
        return value
df2["cuisines"]=df["cuisines"].apply(cuisines_less)
```
